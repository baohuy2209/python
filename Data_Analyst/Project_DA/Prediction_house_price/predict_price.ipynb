###### Display missing values: 

```python
def display_missing_value(df, feature_cols):
    n_rows = df.shape[0] 
    for col in feature_cols: 
        missing_count = df[col].isnull().sum() 
        if missing_count > 0: 
            print(f'Column {col} has {missing_count*100/n_rows : .2f}% missing values')
```

###### Display missing values in each column: 

```python
def display_missing_values_column(df, name):
    n_rows = df.shape[0] 
    
    missing_values = df[name].isnull().sum()
    missing_values_percent = 100 * missing_values / n_rows

    print(f'There are {missing_values} missing values in the {name} column.')
    print(f'{missing_values_percent:.1f}% of the values are missing.')
```
