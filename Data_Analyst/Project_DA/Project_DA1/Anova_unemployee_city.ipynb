## Anova test between unemployee_rate of city in specific vital economic region 

### H0 : µ2 = µ3 = µ4 = µ5 = µ6 = µ7
### H1 : Tồn tại ít nhất một cặp khác nhau.
### Mức ý nghĩa α = 0.05

```python
import pandas as pd 
import numpy as np
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
my_data = pd.read_csv("C:/Users/ADMIN/BAOHUY/Phan_tich_du_lieu/unemployeed_city.csv")
```

```python
my_data 
```

```python
my_data  = my_data .drop([6,7,8,9,10])
my_data 
```

```python
my_data.info()
```

```python
my_data["Đông Nam Bộ"] = my_data["Đông Nam Bộ"].astype("float64")
```

```python
my_data.info()
```

```python
my_data[["Đồng bằng sông Hồng", "Trung du và miền núi phía Bắc", "Bắc Trung Bộ và duyên hải miền Trung", "Tây Nguyên", "Đông Nam Bộ", "Đồng bằng sông Cửu Long"]].describe()
```

```python
from scipy.stats import f_oneway
```

```python
f_stat, p_value = f_oneway(my_data["Đồng bằng sông Hồng"], my_data["Trung du và miền núi phía Bắc"], my_data["Bắc Trung Bộ và duyên hải miền Trung"], my_data["Tây Nguyên"], my_data["Đông Nam Bộ"], my_data["Đồng bằng sông Cửu Long"])
```

```python
print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
if p_value < 0.05:
    print("Bác bỏ H0. Có sự khác biệt ý nghĩa giữa các nhóm.")
else:
    print("Không bác bỏ H0. Không có sự khác biệt ý nghĩa giữa các nhóm.")
```

```python
tukey_data = pd.read_excel("C:/Users/ADMIN/BAOHUY/Phan_tich_du_lieu/unemployee_city_tukey.xlsx")
```

```python
tukey = pairwise_tukeyhsd(endog=tukey_data['unemployee_rate'], groups=tukey_data['area'], alpha=0.05)
```

```python
tukey.plot_simultaneous()
import matplotlib.pyplot as plt
plt.title('Kiểm định Tukey về tỷ lệ thất nghiệp ở thành thị tại các vùng kinh tế trọng điểm')
plt.show()
```
