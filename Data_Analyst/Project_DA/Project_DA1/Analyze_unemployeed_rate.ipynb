## Vẽ biểu đồ sự biến động tỷ lệ thất nghiệp các vùng kinh tế trọng điểm trên cả nước 

```python
import os
file_path = "C:/Users/ADMIN/BAOHUY/Phan_tich_du_lieu/unemployeed_area.csv"
if os.path.exists(file_path):
    unemployeed_area = pd.read_csv(file_path)
else:
    print("Tệp không tồn tại!")
```

```python
unemployeed_area.head()
```

```python
area = ["Cả nước", "Đồng bằng sông Hồng", "Trung du và miền núi phía Bắc", "Bắc Trung Bộ và duyên hải miền Trung", "Tây Nguyên", "Đông Nam Bộ", "Đồng bằng sông Cửu Long"]
```

```python
list_df_area = []
for i in range(len(area)): 
    list_df_area.append(pd.DataFrame({
        'Year': [2018, 2019, 2020, 2021, 2022, 2023], 
        'Value': unemployeed_area[area[i]]
    }))
list_df_area
```

```python
plt.figure(figsize=(10,6))
color = ["blue", "green", "red", "yellow", "black", "pink", "grey"]
for i in range(len(area)):
    plt.plot(list_df_area[i]['Year'], list_df_area[i]['Value'], marker='X', label=area[i], color=color[i])
    # for j in range(len(list_df_area[i])): 
    #     plt.text(list_df_area[i]['Year'][j], list_df_area[i]['Value'][j], str(list_df_area[i]['Value'][j]), ha='center', va='top')
plt.xlabel('Năm')
plt.ylabel(' Tỷ lệ thất nghiệp (%)')
plt.title("Tỷ lệ thất nghiệp các vùng kinh tế trọng điểm trên cả nước (%)")
plt.legend()
```

```python
import seaborn.objects as so 
```

```python
short_area=["Cả nước", "Đồng bằng sông Hồng", "TD và MNPB", "BTB và DHMT", "Tây Nguyên", "Đông Nam Bộ", "ĐBSCL"]
n_rows = len(area)
n_cols = 2 
fig,ax = plt.subplots(n_rows, n_cols, figsize=(10, 20))

for i in range(len(area)): 
    ax_1 = ax[i,0]
    ax_2 = ax[i,1]
    sns.barplot(x=list_df_gender[i]['Year'], y=list_df_gender[i]['Value1'], color="blue", ax=ax_1)
    ax_1.set_title(f"Tỷ lệ thất nghiệp giới Nam trên {short_area[i]} (%)")
    for p in ax_1.patches:
        ax_1.annotate(f'{p.get_height():.2f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), 
                    textcoords='offset points')
    sns.barplot(x=list_df_gender[i]['Year'], y=list_df_gender[i]['Value2'], color="yellow", ax=ax_2)
    ax_2.set_title(f"Tỷ lệ thất nghiệp giới Nữ trên {short_area[i]} (%)")
    for p in ax_2.patches:
        ax_2.annotate(f'{p.get_height():.2f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), 
                    textcoords='offset points')
plt.subplots_adjust(wspace=3, hspace=3)
plt.tight_layout()

```
