## Regression

```python
import pandas as pd 
import numpy as np 
from scipy.stats import linregress
```

```python
amazon_data = pd.read_excel("C:/Users/ADMIN/BAOHUY/Phan_tich_du_lieu/amazon_data2.xlsx")
```

```python
amazon_data = amazon_data.drop(columns=['seller_name.1'])
```

```python
amazon_data.tail(10)
```

```python
initial_price = amazon_data["initial_price"].to_numpy()
```

```python
initial_price 
```

```python
final_price = amazon_data["final_price"].to_numpy()
```

```python
final_price
```

```python
review_count = amazon_data["reviews_count"].to_numpy()
```

```python
review_count
```

```python
alpha = 0.05
```

## Regression between initial price and final price 

```python
slope, intercept, r_value, p_value, std_err = linregress(initial_price, final_price)
```

```python
print(f"Phương trình hồi quy: y = {slope:.2f} * x + {intercept:.2f}")
print(f"Hệ số tương quan r: {r_value:.2f}")
print(f"Giá trị p-value: {p_value:.3f}")
```

## H0: R^2 = 0 
## H1: R^2 != 0 

```python
if p_value < 0.05:
    print("Bác bỏ H0. Tương quan giữa hai biến initial_price và final_price là phù hợp.")
else:
    print("Không bác bỏ H0. Tương quan giữa hai biến initial_price và final_price là không phù hợp.")
```
