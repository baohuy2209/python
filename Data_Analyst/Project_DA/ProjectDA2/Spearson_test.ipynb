# Spearman

```python
import pandas as pd 
import numpy as np 
from scipy.stats import spearmanr
```

```python
amazon_data = pd.read_excel("C:/Users/ADMIN/BAOHUY/Phan_tich_du_lieu/amazon_data2.xlsx")
```

```python
amazon_data = amazon_data.drop(columns=['seller_name.1'])
```

```python
amazon_data.tail(10)
```

```python
rating = amazon_data["rating"].to_numpy()
```

```python
rating 
```

```python
alpha = 0.05
```

## Between rating and initial price

## H0: Hệ số tương quan spearman giữa rating và initial price bằng 0 
## H1: Hệ số tương quan spearman giữa rating và initial price khác 0 

```python
initial_price = amazon_data["initial_price"].to_numpy()
```

```python
initial_price 
```

```python
r, p_value = spearmanr(initial_price , rating )
```

```python
print(f"Hệ số tương quan Pearson: {r}")
print(f"Giá trị p-value: {p_value}")
```

```python
if p_value < 0.05:
    print("Bác bỏ H0. Có mối tương quan ý nghĩa giữa hai biến.")
else:
    print("Không bác bỏ H0. Không có mối tương quan ý nghĩa giữa hai biến.")
```

## Between rating and final price 

## H0: Hệ số tương quan spearman giữa rating và final price bằng 0 
## H1: Hệ số tương quan spearman giữa rating và final price khác 0 

```python
final_price = amazon_data["final_price"].to_numpy()
```

```python
final_price
```

```python
r, p_value = spearmanr(final_price , rating )
```

```python
print(f"Hệ số tương quan Pearson: {r}")
print(f"Giá trị p-value: {p_value}")
```

```python
if p_value < 0.05:
    print("Bác bỏ H0. Có mối tương quan ý nghĩa giữa hai biến.")
else:
    print("Không bác bỏ H0. Không có mối tương quan ý nghĩa giữa hai biến.")
```

## Between initial price and final price 

## H0: Hệ số tương quan spearman giữa initial price và final price bằng 0
