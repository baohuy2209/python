## Wilconxon sign rank cell phone accessories

# H0: med3 = 4.5 
# H1: med3 != 4.5 

```python
import pandas as pd 
import numpy as np 
from scipy.stats import wilcoxon 
```

```python
my_data = pd.read_excel("C:/Users/ADMIN/BAOHUY/Phan_tich_du_lieu/rating_phone_accessories.xlsx")
```

```python
my_data.head()
```

```python
my_data.tail()
```

```python
alpha = 0.05 # mức ý nghĩa 
x = 4.5 #
type_item = "Cell Phone Accessories"
```

```python
transformed_data = [round((float(i) - x),2) for i in my_data["rating"]]
transformed_data
```

```python
stat, p = wilcoxon(transformed_data)
```

```python
print(f"Statistic: {stat}")
print(f"p-value: {p}")
```

```python
if p < alpha: 
    print(f"Bác bỏ giả thuyết H0: rating của ngành hàng {type_item} khác {x}")
else: 
    print(f"Không đủ bằng chứng để bác bỉ giả thuyết H0: rating của ngành hàng {type_item} bằng {x}")
```
