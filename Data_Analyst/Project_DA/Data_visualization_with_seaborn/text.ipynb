```python
dict = {'First Score': [100, 90, np.nan, 95], 
        'Second Score': [30, 45, 56, np.nan], 
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(dict)
df
```

```python
df.isnull().sum()
```

```python
df1 = df.fillna(0)
df1
```

```python
df2 = df.fillna(df.mean())
df2
```

```python
df3 = df.fillna(df.median())
df3
```

```python
df4 = df.fillna(df.mode())
df4
```

```python
df5 = df.fillna(method = "pad")
df5
```

```python
df6=df.fillna(method = "bfill")
df6
```

# Handle duplicate data 

Dữ liệu trùng lặp có thể do sai sót nhập liệu nên xảy ra. Dữ liệu thường trùng ở 2 dòng trở lên.

Ta có thể sử dụng lệnh drop_duplicates() như sau:

```python
df_lap = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df_lap
```

```python
lap1 = df_lap.drop_duplicates()
lap1
```

```python
lap2 = df_lap.drop_duplicates(subset=['brand'])
lap2
```

```python
lap3 = df_lap.drop_duplicates(subset=['brand', 'style'], keep='last')
lap3
```

# Biểu đồ boxplot 

Ta có thể vẽ biểu đồ box plot như sau: 

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue='time')
plt.show()
```

# Biểu đồ violin 

Biểu đồ này giống boxplot nhưng thể hiện thêm phân phối 

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='time')
plt.show()
```

# Biểu đồ pair plot 

Biểu đồ này sẽ vẽ tất cả các cặp biến trong cùng một hình. Cụ thể như sau: 

```python
sns.pairplot(tips)
plt.show()
```

# Strip plot 


Ở các biểu đồ trước, thường ta hiển thị dữ liệu định lượng. Còn đối với dữ liệu định tính thì ta có thể sử dụng Strip plot như sau: 

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', hue = 'time')
plt.show()
```

```python
sns.stripplot(data = tips, y = 'day', x='total_bill', hue='sex')
plt.show()
```

# Swarm Plot 

Biểu đồ này tương tự như trên, nhưng tránh được tính trùng lặp giữa các điểm 

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='time')
```

# Reg plot 

Hàm này được sử dụng để vẽ mô hình hồi quy tuyến tính y~x. Cụ thể như sau: 

```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```

# Resid plot 

Hàm này được sử dụng để vẽ residual của các biến trong mô hình hồi quy tuyến tính 

```python
sns.residplot(x="total_bill", y="tip", data=tips)
```
