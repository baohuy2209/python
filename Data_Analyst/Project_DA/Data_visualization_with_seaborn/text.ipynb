```python
dict = {'First Score': [100, 90, np.nan, 95], 
        'Second Score': [30, 45, 56, np.nan], 
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(dict)
df
```

```python
df.isnull().sum()
```

```python
df1 = df.fillna(0)
df1
```

```python
df2 = df.fillna(df.mean())
df2
```

```python
df3 = df.fillna(df.median())
df3
```

```python
df4 = df.fillna(df.mode())
df4
```

```python
df5 = df.fillna(method = "pad")
df5
```

```python
df6=df.fillna(method = "bfill")
df6
```

# Handle duplicate data 

D liu trng lp c th do sai st nhp liu nn xy ra. D liu thng trng  2 dng tr ln.

Ta c th s dng lnh drop_duplicates() nh sau:

```python
df_lap = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df_lap
```

```python
lap1 = df_lap.drop_duplicates()
lap1
```

```python
lap2 = df_lap.drop_duplicates(subset=['brand'])
lap2
```

```python
lap3 = df_lap.drop_duplicates(subset=['brand', 'style'], keep='last')
lap3
```
