## 1. Pyplot API & Object-Oriented (OO) API

. Pyplot API -> Quickly 
. OO API -> Advanced

###### Pyplot API 

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)")
plt.plot(x, np.cos(x), color = "red", label = "cos(x)")

plt.title("Sin(x) and Cos(x)")
plt.xlabel("x")
plt.ylabel("F(x)")

#plt.xlim([0,4])
#plt.ylim([-0.7, 0.6])
#plt.axis([0.4, -0.7, 0.6])

plt.axis("equal")
plt.legend()
```

##### OO API 

```python
x = [1,2,3,4]
y = [11,22,33,44]

fig, ax = plt.subplots(figsize = (5,5)) 
ax.plot(x,y)

ax.set(title = "A simple plot", 
      xlabel = "x-axis", 
      ylabel = "y-axis")
```

### 3. Bar 
* Vertical 
* Horizontal 

```python
solf_drink_prices = {
    "coke" : 10, 
    "Pessi" : 12, 
    "Sprite" : 8
}

fig,ax = plt.subplots() 
ax.bar(solf_drink_prices.keys(), solf_drink_prices.values())
ax.set(
    title = "Bach hoa Xanh's soft drink prices", 
    ylabel = "Price ($)"
)
```

```python
fig,ax = plt.subplots() 
ax.barh(list(solf_drink_prices.keys()), list(solf_drink_prices.values()))
```

### Subplots 
* Multiple plots one figures

```python
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize =(10,5))

ax1.plot(x, x/2)
ax1.plot(x, x*2)

ax2.scatter(np.random.random(10), np.random.random(10))

ax3.bar(solf_drink_prices.keys(), solf_drink_prices.values())

ax4.hist(student_height, bins = 30)
```

```python
fig, ax = plt.subplots(nrows = 2,
                        ncols = 2, 
                      figsize = (10,5))

ax[0,0].plot(x, x/2)
ax[0,0].plot(x, x*2)

ax[0,1].scatter(np.random.random(10), np.random.random(10))

ax[1,0].bar(solf_drink_prices.keys(), solf_drink_prices.values())

ax[1,1].hist(student_height, bins = 30)
```

### Project : Data Visualization - California cities

```python
df = pd.read_csv("C:/Users/Admin/data/california_cities.csv")

df["area_total_km2"].sort_values(ascending = False)
```

```python
df.head() 
```

```python
# Extract latd and longd 
lat, lon = df["latd"], df["longd"]
population, area = df["population_total"], df["area_total_sq_mi"]

plt.scatter(lon, lat, 
           c = np.log10(population), cmap = "viridis",
           s = area, linewidths = 0 , alpha = 0.5)
plt.xlabel("Longtitude")
plt.ylabel("Lattitude")
plt.axis("equal")
plt.colorbar(label = 'log$_{10}${population}')

plt.title('California Cities : Population and Area Distribution')


area_range = [50, 100, 300, 500]
for area in area_range: 
    plt.scatter([], [], s = area, c = 'k',
                alpha = 0.5,
               label = str(area)+'km$^2$')
    
plt.legend(scatterpoints = 1, labelspacing = 1, title = "City Area")
```
